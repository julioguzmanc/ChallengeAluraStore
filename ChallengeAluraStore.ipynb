# 2. Ventas por categoría

```python
# Función para contar ventas por categoría
def ventas_por_categoria(df, nombre_tienda):
    conteo = df.groupby("Categoría del Producto")["Producto"].count().reset_index()
    conteo.columns = ["Categoría del Producto", "Cantidad de Ventas"]
    conteo = conteo.sort_values(by="Cantidad de Ventas", ascending=False)
    print(f"\nCategorías más populares en {nombre_tienda}:")
    print(conteo.to_string(index=False))

# Mostrar resultados para cada tienda en orden de categorias más populares
ventas_por_categoria(tienda, "Tienda 1")
ventas_por_categoria(tienda2, "Tienda 2")
ventas_por_categoria(tienda3, "Tienda 3")
ventas_por_categoria(tienda4, "Tienda 4")
```

```python
# ---------------- 2. Gráfico Circular - Distribución de Categorías ----------------
categorias_t1 = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(categorias_t1, labels=categorias_t1.index, autopct="%1.1f%%", startangle=90, colors=plt.cm.Paired.colors)
plt.title("Distribución de Categorías - Tienda 1")
plt.show()
```

```python
# ---------------- 1. Gráfico de Barras - Ingresos por tienda ----------------
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color=["#4CAF50", "#2196F3", "#FFC107", "#FF5722"])
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tienda")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# ---------------- 2. Gráficos combinados para cada tienda ----------------
fig, axes = plt.subplots(4, 2, figsize=(12, 20))
fig.suptitle("Análisis por Tienda", fontsize=16, fontweight="bold")

tiendas_datos = [
    ("Tienda 1", tienda),
    ("Tienda 2", tienda2),
    ("Tienda 3", tienda3),
    ("Tienda 4", tienda4)
]

for i, (nombre, df) in enumerate(tiendas_datos):
    # Pie chart - Categorías
    categorias = df["Categoría del Producto"].value_counts()
    axes[i, 0].pie(categorias, labels=categorias.index, autopct="%1.1f%%", startangle=90, colors=plt.cm.Paired.colors)
    axes[i, 0].set_title(f"{nombre} - Distribución de Categorías")

    # Dispersión - Precio vs Calificación
    axes[i, 1].scatter(df["Precio"], df["Calificación"], alpha=0.6, color="purple", edgecolors="black")
    axes[i, 1].set_title(f"{nombre} - Precio vs Calificación")
    axes[i, 1].set_xlabel("Precio ($)")
    axes[i, 1].set_ylabel("Calificación (1-5)")
    axes[i, 1].grid(True, linestyle="--", alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs de las bases de datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar datos
tiendas = {
    "Tienda 1": pd.read_csv(url),
    "Tienda 2": pd.read_csv(url2),
    "Tienda 3": pd.read_csv(url3),
    "Tienda 4": pd.read_csv(url4)
}

# Analizar cada tienda
for nombre, df in tiendas.items():
    conteo = df.groupby("Producto")["Producto"].count().sort_values(ascending=False)

    producto_mas_vendido = conteo.idxmax()
    ventas_mas_vendido = conteo.max()

    producto_menos_vendido = conteo.idxmin()
    ventas_menos_vendido = conteo.min()

    print(f"\n📊 {nombre}")
    print(f"   🥇 Más vendido: {producto_mas_vendido} ({ventas_mas_vendido} ventas)")
    print(f"   🥉 Menos vendido: {producto_menos_vendido} ({ventas_menos_vendido} ventas)")

    # Visualizar
    plt.figure(figsize=(16,4))
    conteo.plot(kind="bar", color="skyblue", edgecolor="black")
    plt.title(f"Ventas por producto - {nombre}")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad de Ventas")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
